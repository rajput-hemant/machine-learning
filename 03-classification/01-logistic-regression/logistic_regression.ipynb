# Logistic Regression


**Q. What is Logistic Regression?**

- Logistic Regression is a classification algorithm used to predict the probability of a categorical dependent variable from a set of independent variables.
- It is a supervised learning algorithm that uses a logistic function to model a binary dependent variable.
- It is used to predict a binary outcome (1 / 0, Yes / No, True / False) given a set of independent variables.
- It is an extension of linear regression that is used when the dependent variable is categorical.
- In other words, the logistic regression model predicts P(Y=1) as a function of X.


## Importing the libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset


```python
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)  # split data into train and test with 25% test size and random state 0 to get same result
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling


```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(
    X_train
)  # fit i.e. calculate mean and standard deviation and then transform i.e. standardize the data
X_test = sc.transform(
    X_test
)  # only transform i.e. standardize the data, no need to fit again as we already fit the data
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the Logistic Regression model on the Training set


```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)  # random state 0 to get same result
classifier.fit(X_train, y_train)  # train the classifier on the training set
```

## Predicting a new result


```python
print(
    classifier.predict(
        sc.transform(
            [[30, 87000]]
        )  # [[age, salary]] to predict if the person will buy the product or not
    )  # standardize the input data and then predict the output
)
```

## Predicting the Test set results


```python
y_pred = classifier.predict(X_test)
print(
    np.concatenate(
        (
            y_pred.reshape(len(y_pred), 1),  # reshape vector to be horizontal
            y_test.reshape(len(y_test), 1),
        ),
        1,
    )
)
```

## Making the Confusion Matrix


**Q. What is a Confusion Matrix?**

- A confusion matrix is a table that is often used to describe the performance of a classification model (or “classifier”) on a set of test data for which the true values are known.
- Each row of the matrix represents the instances in a predicted class while each column represents the instances in an actual class.
- The name stems from the fact that it makes it easy to see if the system is confusing two classes (i.e. commonly mislabeling one as another).


```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
#  [[65  3] 65 correct predictions for 0 and 3 incorrect predictions for 0
#  [ 8 24]] 8 incorrect predictions for 1 and 24 correct predictions for 1
# 0 means did not buy the product and 1 means bought the product

accuracy_score(y_test, y_pred)
```

## Visualising the Training set results


```python
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=0.40),
    np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=0.40),
)  # the lower the step the more accurate the graph will be, but it will take more time and memory to compute
plt.contourf(
    X1,
    X2,
    classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(
        X1.shape
    ),
    alpha=0.75,
    cmap=ListedColormap(("red", "green")),
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        X_set[y_set == j, 0],
        X_set[y_set == j, 1],
        c=ListedColormap(("red", "green"))(i),
        label=j,
    )
plt.title("Logistic Regression (Training set)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```

## Visualising the Test set results


```python
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=0.40),
    np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=0.40),
)
plt.contourf(
    X1,
    X2,
    classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(
        X1.shape
    ),
    alpha=0.75,
    cmap=ListedColormap(("red", "green")),
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        X_set[y_set == j, 0],
        X_set[y_set == j, 1],
        c=ListedColormap(("red", "green"))(i),
        label=j,
    )
plt.title("Logistic Regression (Test set)")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.legend()
plt.show()
```
